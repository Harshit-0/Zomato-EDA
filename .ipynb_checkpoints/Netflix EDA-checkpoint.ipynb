## ðŸ“Œ Exploratory Data Analysis

```python
#1. Movie watchers VS TV Watchers
plt.figure(figsize=(6, 4))
sns.set_style("whitegrid")

ax = sns.countplot(data=df, x='type', palette='Set2')
plt.title("Content Type Distribution on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Count")

# Add labels
counts = df['type'].value_counts()
for i, v in enumerate(counts):
    ax.text(i, v + 50, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Netflix has more *Movies* than TV Shows, indicating a film-heavy catalog. However, TV shows form a significant chunk, reflecting a shift towards binge-worthy content.

```python
#2. Which Year has the most releases
# Create a new column for year
df['year_added'] = df['date_added'].dt.year

# Drop NaNs just for this plot
yearly = df['year_added'].dropna().astype(int).value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly.index, y=yearly.values, marker='o', color='crimson')
plt.title("Year-wise Content Addition on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Netflix saw a significant rise in content additions from *2016* to *2019*, peaking around *2019*. This growth aligns with its global expansion and increased investment in original content during that period.

```python
#3. Top 10 countries producing Nettflix Content
# Drop NaNs and split multiple countries
country_split = df['country'].dropna().str.split(', ')

# Flatten using explode (Pandas-only)
all_countries = country_split.explode()

# Count and get top 10
top_countries = all_countries.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

The *United States* is by far the top contributor to Netflix content, followed by *India*, *United Kingdom*, and *Canada*. This highlights Netflixâ€™s production and licensing priorities worldwide.

```python
#4. Most common genres
# Drop missing values and split genres
genre_split = df['listed_in'].dropna().str.split(', ')

# Flatten using explode
all_genres = genre_split.explode()

# Count top genres
top_genres = all_genres.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

## ðŸ“Œ Insight

*Dramas* are the most common genre on Netflix, followed by *Comedies* and *Documentaries*. This suggests a preference for story-driven and informative content among viewers.

```python
#5. Top 10 Directors with most Content on Netflix
# Clean the director column
clean_directors = df['director'].dropna()
clean_directors = clean_directors[clean_directors.str.strip().str.lower() != 'nan']
clean_directors = clean_directors[clean_directors.str.strip() != '']

# Get top 10 directors
top_directors = clean_directors.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='mako')
plt.title("Top 10 Directors by Number of Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

A few directors stand out as major contributors to Netflixâ€™s catalog. Some of them may specialize in regional cinema or frequently collaborate with Netflix for original content.

```python
#6.Movie Durations VS TV Show Seasons
# Separate Movies and TV Shows
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show'].copy()

# Clean duration for Movies (remove 'min' and convert to int)
movies['duration_int'] = movies['duration'].str.replace(' min', '').astype(float)

# Clean duration for TV Shows (remove 'Season(s)' and convert to int)
tv_shows['duration_int'] = tv_shows['duration'].str.replace(' Season', '')
tv_shows['duration_int'] = tv_shows['duration_int'].str.replace('s', '').astype(float)

# Plot side-by-side
plt.figure(figsize=(12, 5))

# Movie durations
plt.subplot(1, 2, 1)
sns.histplot(movies['duration_int'], bins=30, color='skyblue', kde=True)
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")

# TV Show seasons
plt.subplot(1, 2, 2)
sns.histplot(tv_shows['duration_int'], bins=10, color='salmon', kde=False)
plt.title("TV Show Season Distribution")
plt.xlabel("Number of Seasons")

plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Netflix Movies mostly range between *60 to 120 minutes*, while most TV Shows have *1 to 2 seasons*, showing a trend toward shorter, bingeable series.

```python
#7.Top 10 Most Frequent Actors on Netflix
# Drop nulls and remove invalid strings
clean_cast = df['cast'].dropna()
clean_cast = clean_cast[clean_cast.str.strip().str.lower() != 'nan']
clean_cast = clean_cast[clean_cast.str.strip() != '']

# Split multiple actors and flatten
cast_split = clean_cast.str.split(', ')
all_actors = cast_split.explode()

# Count top 10 actors
top_actors = all_actors.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_actors.values, y=top_actors.index, palette='cubehelix')
plt.title("Top 10 Most Frequent Actors on Netflix")
plt.xlabel("Number of Appearances")
plt.ylabel("Actor")
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Some actors are featured in multiple Netflix projects, often across regional or international titles â€” showing recurring collaborations or high audience demand.

```python
#9.Content Rating Distribution (Target Audience)
# Drop NaNs in 'rating'
ratings_clean = df.dropna(subset=['rating'])

# Plot
plt.figure(figsize=(10,6))
sns.countplot(data=ratings_clean, x='rating', order=ratings_clean['rating'].value_counts().index[:10], palette='viridis')
plt.title('Distribution of Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Helps to show what kind of audience Netflix targets (e.g. TV-MA,R,Etc.)

```python
#10.Heatmap of content added by Month and Year
# Drop NaNs in 'date_added'
heatmap_data = df.dropna(subset=['date_added'])

# Extract month & year
heatmap_data['month'] = heatmap_data['date_added'].dt.month
heatmap_data['year'] = heatmap_data['date_added'].dt.year

# Create pivot
pivot_table = heatmap_data.pivot_table(index='month', columns='year', aggfunc='size', fill_value=0)

# Plot
plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5, linecolor='gray')
plt.title('Netflix Titles Added by Month & Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

### ðŸ“Œ Insight

Highlights seasonal and yearly content trends
